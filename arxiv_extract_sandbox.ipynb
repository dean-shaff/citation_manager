```python
from bs4 import BeautifulSoup

with open('example.html','r') as example:
    html = example.read() 

soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
import urllib
for item in soup.find_all('meta'):
    pass
    #     print(item)
#     print(item['content'], item['name'])
#     print(item.get('citation_author'))
for item in soup.find_all('a'):
    if item.text == "NASA ADS":
        new_url = item.attrs['href']
print(new_url)
html_ads = urllib.urlopen(new_url).read()
```

```python
soup_ads = BeautifulSoup(html_ads)
for item in soup_ads.find_all('a'):
    if ("Bibtex" in item.text):
        bibtex_url = item.attrs['href']
bibtex_html = urllib.urlopen(bibtex_url).read()
```

```python
for i, char in enumerate(bibtex_html):
    if char == "@":
        bibtex_entry = bibtex_html[i:]
        break
print(bibtex_entry)
```
